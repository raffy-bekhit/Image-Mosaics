```python
def transform(point,h):
	h_2d = np.zeros((3,3));
	for i in range(0,3):
		for j in range(0,3):
			if 3*i+j < 8:
				h_2d[i][j] = h[3*i+j];
	h_2d[2][2] = 1;
#	print('h_2d ' + str(h_2d));
	ret = np.dot(h_2d, point);
	ret[0] = ret[0]/ret[2];
	ret[1] = ret[1]/ret[2];
	ret[2] = 1;
	return ret;
```
